```python
# ===========================================
# Corrected Lightweight MAE vs PCA ‚Äî full notebook
# ===========================================
# Run in Colab / local. Assumes dataset layout like your previous run (RBG/train, RBG/val, RBG/test).
# Change paths or dataloaders if necessary.

# 0Ô∏è‚É£ Install packages (uncomment if needed in Colab)
# !pip install --quiet torch torchvision timm einops scikit-image scikit-learn matplotlib kagglehub tqdm

# 1Ô∏è‚É£ Imports & Setup
import os, math, random, time
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as T
from torchvision.datasets import ImageFolder
from sklearn.decomposition import PCA
from skimage.metrics import peak_signal_noise_ratio as compare_psnr
from skimage.metrics import structural_similarity as compare_ssim
from einops import rearrange

# Device & seed
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Using device:", device)
SEED = 42
torch.manual_seed(SEED); np.random.seed(SEED); random.seed(SEED)

# --------------- metrics helpers ---------------
def compute_psnr(a, b): return compare_psnr(a, b, data_range=1.0)
def compute_ssim(a, b): return compare_ssim(a, b, channel_axis=2, data_range=1.0)
def compute_sam(a, b, eps=1e-8):
    v1 = a.reshape(-1, a.shape[2])
    v2 = b.reshape(-1, b.shape[2])
    dot = (v1 * v2).sum(axis=1)
    denom = (np.linalg.norm(v1, axis=1) * np.linalg.norm(v2, axis=1)) + eps
    return np.mean(np.arccos(np.clip(dot / denom, -1.0, 1.0)))

def show_images(images, titles=None, figsize=(12,6)):
    plt.figure(figsize=figsize)
    for i, img in enumerate(images):
        plt.subplot(1, len(images), i+1)
        plt.imshow(np.clip(img,0,1))
        if titles: plt.title(titles[i])
        plt.axis('off')
    plt.show()

# 2Ô∏è‚É£ Load Dataset via kagglehub (or local)
# If you already have the dataset path, set data_root accordingly.
try:
    import kagglehub
    path = kagglehub.dataset_download("pranjallk1995/rgbeurosat")
    data_root = os.path.join(path, "RBG")
    print("Dataset path via kagglehub:", data_root)
except Exception as e:
    # If kagglehub not available or you have local copy, adjust this
    data_root = "/root/.cache/kagglehub/datasets/pranjallk1995/rgbeurosat/versions/1/RBG"
    print("Using fallback dataset path (adjust if wrong):", data_root)

img_size = 128
transform = T.Compose([T.Resize((img_size,img_size)), T.ToTensor()])

# Load train and val from dataset folders (expecting 'train' & 'val' subfolders)
train_dir = os.path.join(data_root, "train")
val_dir = os.path.join(data_root, "val")
test_dir = os.path.join(data_root, "test")  # optional

if not os.path.exists(train_dir):
    # if dataset flat, use the single folder (previously you had ImageFolder(root=data_dir))
    data_dir = data_root
    dataset = ImageFolder(root=data_dir, transform=transform)
    # do split
    n = len(dataset)
    nval = int(0.05 * n)
    ntrain = n - nval
    train_set, val_set = random_split(dataset, [ntrain, nval], generator=torch.Generator().manual_seed(SEED))
else:
    train_set = ImageFolder(root=train_dir, transform=transform)
    val_set = ImageFolder(root=val_dir, transform=transform)

batch_size = 64
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, drop_last=True, num_workers=4)
val_loader = DataLoader(val_set, batch_size=32, shuffle=False, drop_last=False, num_workers=4)
print("Train images:", len(train_set), "Val images:", len(val_set))

# 3Ô∏è‚É£ Model: PatchEmbed + MAE (batch_first=True, returns mask info)
class PatchEmbed(nn.Module):
    def __init__(self, img_size=128, patch_size=16, in_ch=3, embed_dim=256):
        super().__init__()
        self.patch_size = patch_size
        self.proj = nn.Conv2d(in_ch, embed_dim, kernel_size=patch_size, stride=patch_size)
        self.img_size = img_size
        self.num_patches = (img_size // patch_size) ** 2

    def forward(self, x):
        # x: (B,3,H,W)
        x = self.proj(x)          # (B,embed,h',w')
        return rearrange(x, 'b c h w -> b (h w) c')

class MAELightCorrected(nn.Module):
    def __init__(self, img_size=128, patch_size=16, embed_dim=256, depth=6, decoder_depth=4, mask_ratio=0.75):
        super().__init__()
        self.patch_size = patch_size
        self.embed_dim = embed_dim
        self.num_patches = (img_size // patch_size) ** 2
        self.mask_ratio = mask_ratio

        self.patch_embed = PatchEmbed(img_size, patch_size, 3, embed_dim)
        self.pos_embed = nn.Parameter(torch.randn(1, self.num_patches, embed_dim) * 0.02)
        self.mask_token = nn.Parameter(torch.zeros(1, 1, embed_dim))
        nn.init.normal_(self.mask_token, std=0.02)

        # use batch_first=True to simplify
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=8, dim_feedforward=embed_dim*2, batch_first=True)
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=depth)

        decoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=8, dim_feedforward=embed_dim*2, batch_first=True)
        self.decoder = nn.TransformerEncoder(decoder_layer, num_layers=decoder_depth)

        self.decoder_pred = nn.Linear(embed_dim, patch_size*patch_size*3)

        # Optional layernorms
        self.norm = nn.LayerNorm(embed_dim)

    def forward(self, imgs, return_mask_info=True):
        """
        Returns:
          pred: (B, N, patch_dim) decoded predictions for all patches (both visible & masked)
          latent: encoded visible tokens (B, num_keep, D)
          ids_keep, ids_restore, num_keep
        """
        B = imgs.shape[0]
        x = self.patch_embed(imgs) + self.pos_embed  # (B,N,D)
        N = x.shape[1]
        D = x.shape[2]

        # Create random mask per sample
        num_keep = int(N * (1 - self.mask_ratio))
        noise = torch.rand(B, N, device=imgs.device)
        ids_shuffle = torch.argsort(noise, dim=1)        # ascend -> shuffled indices
        ids_restore = torch.argsort(ids_shuffle, dim=1) # to reconstruct original order
        ids_keep = ids_shuffle[:, :num_keep]            # indices to keep

        # Gather visible patches
        ids_keep_exp = ids_keep.unsqueeze(-1).repeat(1,1,D)
        x_keep = torch.gather(x, dim=1, index=ids_keep_exp)  # (B, num_keep, D)

        # Encode visible tokens
        latent = self.encoder(self.norm(x_keep))  # (B, num_keep, D)

        # prepare mask tokens
        mask_tokens = self.mask_token.repeat(B, N - num_keep, 1)
        x_full = torch.cat([latent, mask_tokens], dim=1)  # (B, N, D) but in shuffled order
        # restore to original patch order
        ids_restore_exp = ids_restore.unsqueeze(-1).repeat(1,1,D)
        x_full = torch.gather(x_full, dim=1, index=ids_restore_exp)  # (B, N, D)
        x_full = x_full + self.pos_embed

        # decode
        dec = self.decoder(self.norm(x_full))  # (B,N,D)
        pred = self.decoder_pred(dec)          # (B,N,patch_dim)

        if return_mask_info:
            return pred, latent, ids_keep, ids_restore, num_keep
        else:
            return pred, latent

# 4Ô∏è‚É£ Helpers: patchify & unpatchify for targets
def img_to_patches(imgs, patch):
    # imgs: (B,3,H,W)
    B = imgs.shape[0]
    return rearrange(imgs, 'b c (h p1) (w p2) -> b (h w) (c p1 p2)', p1=patch, p2=patch)

def patches_to_img(patches, patch, nside):
    # patches: (B, N, c*patch*patch)
    B, N, P = patches.shape
    c = 3
    return rearrange(patches, 'b (h w) (c p1 p2) -> b c (h p1) (w p2)', h=nside, w=nside, c=c, p1=patch, p2=patch)

# 5Ô∏è‚É£ Training loop (masked-patch loss + fixed validation)
def eval_model(model, loader, device):
    model.eval()
    ps_list, ss_list, sam_list = [], [], []
    with torch.no_grad():
        for imgs, _ in loader:
            imgs = imgs.to(device)
            pred, _ = model(imgs, return_mask_info=False)  # ‚úÖ fixed unpacking
            B, N, patch_dim = pred.shape
            nside = int(math.sqrt(N))
            recon = patches_to_img(pred, patch=model.patch_size, nside=nside)  # (B,3,H,W)
            for i in range(B):
                orig = imgs[i].permute(1,2,0).cpu().numpy()
                rec = recon[i].permute(1,2,0).cpu().numpy()
                rec = np.clip(rec, 0.0, 1.0)
                ps_list.append(compute_psnr(orig, rec))
                ss_list.append(compute_ssim(orig, rec))
                sam_list.append(compute_sam(orig, rec))
    return float(np.mean(ps_list)), float(np.mean(ss_list)), float(np.mean(sam_list))


def train_mae(model, train_loader, val_loader, epochs=50, lr=1e-4, device='cpu', save_path=None):
    model = model.to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()
    hist = {'loss': [], 'psnr': [], 'ssim': [], 'sam': []}

    for ep in range(1, epochs+1):
        model.train()
        total_loss = 0.0
        batches = 0
        pbar = tqdm(train_loader, desc=f"Epoch {ep}/{epochs}")
        for imgs, _ in pbar:
            imgs = imgs.to(device)
            pred, latent, ids_keep, ids_restore, num_keep = model(imgs, return_mask_info=True)
            # prepare target patches (in original order)
            patch = model.patch_size
            img_patches = img_to_patches(imgs, patch)  # (B, N, patch_dim)
            # masked indices (per-sample)
            ids_mask = ids_restore[:, num_keep:]  # (B, N_mask)
            # gather masked predictions and masked targets
            # expand indices for gather
            pred_masked = torch.gather(pred, dim=1, index=ids_mask.unsqueeze(-1).repeat(1,1,pred.shape[2]))
            target_masked = torch.gather(img_patches, dim=1, index=ids_mask.unsqueeze(-1).repeat(1,1,img_patches.shape[2]))
            loss = loss_fn(pred_masked, target_masked)

            opt.zero_grad()
            loss.backward()
            opt.step()

            total_loss += loss.item()
            batches += 1
            pbar.set_postfix({'loss': f"{total_loss/batches:.6f}"})

        avg_loss = total_loss / max(1, batches)
        # Validation (average)
        ps, ss, sam = eval_model(model, val_loader, device)
        hist['loss'].append(avg_loss)
        hist['psnr'].append(ps)
        hist['ssim'].append(ss)
        hist['sam'].append(sam)
        print(f"Epoch {ep} done. Loss={avg_loss:.6f} | VAL PSNR={ps:.3f} | VAL SSIM={ss:.3f} | VAL SAM={sam:.5f}")

        if save_path:
            torch.save(model.state_dict(), save_path)

    return model, hist

# 6Ô∏è‚É£ Create & train model for 50 epochs
model = MAELightCorrected(img_size=img_size, patch_size=16, embed_dim=256, depth=6, decoder_depth=4, mask_ratio=0.75)
epochs = 50
model, hist = train_mae(model, train_loader, val_loader, epochs=epochs, lr=1e-4, device=device, save_path="mae_light_corrected.pth")

# 7Ô∏è‚É£ Plot training curves
plt.figure(figsize=(10,4))
plt.plot(hist['loss'], label='Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Training Loss Curve")
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(10,4))
plt.plot(hist['psnr'], label='PSNR')
plt.plot(hist['ssim'], label='SSIM')
plt.xlabel("Epoch"); plt.ylabel("Value"); plt.title("Validation PSNR/SSIM")
plt.legend(); plt.grid(True)
plt.show()

# 8Ô∏è‚É£ Evaluate on one sample image and compare with PCA with matched byte budget
# Pick one validation image
model.eval()
with torch.no_grad():
    sample_img, _ = val_set[0]
    test_img = sample_img.unsqueeze(0).to(device)
    pred, latent, _, _, _ = model(test_img, return_mask_info=True)
    # Reconstruct image from pred
    B, N, patch_dim = pred.shape
    nside = int(math.sqrt(N))
    recon = patches_to_img(pred, patch=model.patch_size, nside=nside)
    recon_np = recon[0].permute(1,2,0).cpu().numpy()
    orig_np = test_img[0].permute(1,2,0).cpu().numpy()
    recon_np = np.clip(recon_np,0,1)

# Show images
show_images([orig_np, recon_np], ["Original (val sample)", "MAE Reconstructed (sample)"])

print("MAE sample metrics:")
print("PSNR:", compute_psnr(orig_np, recon_np))
print("SSIM:", compute_ssim(orig_np, recon_np))
print("SAM:", compute_sam(orig_np, recon_np))

# 9Ô∏è‚É£ Byte accounting for MAE (single-sample latent stored as float32)
latent_cpu = latent.detach().cpu()
dtype_bytes = 4 if latent_cpu.dtype == torch.float32 else 2 if latent_cpu.dtype==torch.float16 else 8
mae_bytes = latent_cpu.numel() * dtype_bytes
print("MAE latent bytes (single sample, float32):", mae_bytes)

# 10Ô∏è‚É£ PCA: choose k so that PCA bytes ‚âà mae_bytes (fair comparison)
# Build pixels of the chosen sample
pixels = orig_np.reshape(-1,3)  # (n_pixels, 3)
n_pixels = pixels.shape[0]
# We're storing PCA as: components (k x 3), coefficients (n_pixels x k), mean (3)
# Bytes for float32:
def pca_bytes_for_k(k, dtype_bytes=4):
    comps = k * 3
    coeffs = n_pixels * k
    mean = 3
    return dtype_bytes * (comps + coeffs + mean)

# find k that makes PCA bytes roughly equal to mae_bytes (or nearest)
max_k = min(512, n_pixels)  # reasonable upper bound
best_k = 1
best_diff = float('inf')
for k in range(1, max_k+1):
    b = pca_bytes_for_k(k, dtype_bytes)
    diff = abs(b - mae_bytes)
    if diff < best_diff:
        best_diff = diff
        best_k = k

print("Matched PCA k:", best_k, "PCA bytes approx:", pca_bytes_for_k(best_k, dtype_bytes), "MAE bytes:", mae_bytes)

# fit PCA with best_k
pca = PCA(n_components=best_k, random_state=SEED)
coeffs = pca.fit_transform(pixels)  # (n_pixels, k)
recon_pca = pca.inverse_transform(coeffs).reshape(img_size, img_size, 3)
recon_pca = np.clip(recon_pca, 0.0, 1.0)

psnr_pca = compute_psnr(orig_np, recon_pca)
ssim_pca = compute_ssim(orig_np, recon_pca)
sam_pca = compute_sam(orig_np, recon_pca)

print("\n--- Comparison (single sample) ---")
print(f"PCA (k={best_k}) bytes ‚âà {pca_bytes_for_k(best_k, dtype_bytes)} -> PSNR={psnr_pca:.3f}, SSIM={ssim_pca:.3f}, SAM={sam_pca:.5f}")
print(f"MAE -> PSNR={compute_psnr(orig_np, recon_np):.3f}, SSIM={compute_ssim(orig_np, recon_np):.3f}, SAM={compute_sam(orig_np, recon_np):.5f}")

# 11Ô∏è‚É£ Visualize PCA vs MAE and compression ratios
show_images([orig_np, recon_pca, recon_np], ["Original", f"PCA k={best_k}", "MAE"])

labels = ['PCA', 'MAE']
psnr_vals = [psnr_pca, compute_psnr(orig_np, recon_np)]
ssim_vals = [ssim_pca, compute_ssim(orig_np, recon_np)]
bytes_vals = [pca_bytes_for_k(best_k, dtype_bytes), mae_bytes]
comp_ratio = [(img_size*img_size*3)/b for b in bytes_vals]

plt.figure(figsize=(8,4)); plt.bar(labels, psnr_vals); plt.title("PSNR Comparison"); plt.ylabel("PSNR"); plt.grid(True); plt.show()
plt.figure(figsize=(8,4)); plt.bar(labels, ssim_vals); plt.title("SSIM Comparison"); plt.ylabel("SSIM"); plt.grid(True); plt.show()
plt.figure(figsize=(8,4)); plt.bar(labels, comp_ratio); plt.title("Compression Ratio (orig / compressed)"); plt.ylabel("Ratio"); plt.grid(True); plt.show()

# 12Ô∏è‚É£ Save final model & history
torch.save(model.state_dict(), "mae_light_corrected_final.pth")
import json
with open("mae_hist.json","w") as f:
    json.dump({k:list(map(float,v)) for k,v in hist.items()}, f)

print("Done. Model and history saved.")
```
